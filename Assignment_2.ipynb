**1. Numpy:**

Using NumPy create random vector of size 15 having only Integers in the range 1-20.

1.1. Reshape the array to 3 by 5

1.2. Print array shape.

1.3. Replace the max in each row by 0

```python
import numpy
#1.1 creating a random vector of range 1-20 and size = 15

ran_vec = numpy.random.randint(1, 20, size = 15).reshape(3,5)
print(ran_vec)
```

```python
#1.2 Priniting th e array size
print("Array-Shape: ", ran_vec.shape)
```

```python
#1.3 Repalcing the max 
Max_val = numpy.amax(ran_vec, axis = 1)

Max_val = Max_val[: , None]

Updated_Arr = numpy.where(ran_vec == Max_val, 0, ran_vec)

print(Updated_Arr)
```

**3. Matplotlib**

1. Write a Python programming to create a below chart of the popularity of programming Languages.
2. Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
%matplotlib inline

from matplotlib import pyplot as plot


lang = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b']

explode = [0.1,0,0,0,0,0]


plot.pie(popularity,labels=lang,colors=colors,explode=explode,startangle=130,
       wedgeprops = {'edgecolor':'black'},autopct='%1.2f%%')




plot.show()
```
